```python
from bs4 import BeautifulSoup
import time, os
import random
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
#import functions for web scraping
from scraper import money_to_int, size_to_float, get_details_value, get_bags_dict
```

```python
chromedriver = "/Applications/chromedriver" # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedriver
```

```python
driver = webdriver.Chrome(chromedriver)
```

```python
#scrape all bags URLs
rows=[]
for i in range(1, 30, 1):
    url_page = 'https://shop.rebag.com/collections/all-bags?page=' + str(i) + '&_=pf&pf_st_availability_hidden=true&pf_t_categories=bc-filter-Bags'
    driver.get(url_page)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    rows_one_page = soup.find_all(class_='product-caption') 
    rows.append(rows_one_page)
    time.sleep(.5+2*random.random())
```

```python
links_bags = []
for row in rows:
    for r in row:
        links_bags.append(r.find('a')['href'])
```

```python
# scrape each URL for bag details
bags_page_info_list = []
for link in links_bags:
    bags_page_info_list.append(get_bags_dict(link, driver))
    time.sleep(5+2*random.random())
```

```python
 #convert list of dict to df
bags_page_info = pd.DataFrame(bags_page_info_list) 
bags_page_info.set_index('Title', inplace=True)

bags_page_info
```
